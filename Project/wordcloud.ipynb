```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from os import path, getcwd
from PIL import Image
```

```python
data_folder = "./data/"
```

```python
france = pd.read_csv(data_folder + "france_data.csv", sep=',', dtype=object)
us = pd.read_csv(data_folder + "us_data.csv", sep=',', dtype=object)
us_mask = np.array(Image.open("img/us_mask.png"))
fr_mask = np.array(Image.open("img/fr_mask.png"))
```

# By product name

```python
text = us['product_name'].copy().sample(60000).astype(str).agg(sum)

# Display the generated image:
wordcloud = WordCloud(max_words=150, background_color="white", mask = us_mask, colormap="Reds").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file('img/us.png')
```

```python
text = france['product_name'].sample(60000).astype(str).agg(sum)
stopwords = set(STOPWORDS)
stopwords.update(["au", "de", "la", "aux", "en", "du", "et", "sans","des","le"])

# Generate a word cloud image
wordcloud = WordCloud(max_words=150, stopwords=stopwords, background_color="white", mask = fr_mask, colormap="Blues")
wordcloud.generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file('img/france.png')
```

# By categories

```python
text = us['categories_en'].copy().sample(6000).astype(str).agg(sum)
us_mask = np.array(Image.open("img/us_mask.png"))
stopwords = set(STOPWORDS)
stopwords.update(["based","foods","beverages","products"])

# Display the generated image:
wordcloud = WordCloud(max_words=150, background_color="white", mask = us_mask, colormap="Paired", stopwords = stopwords)
wordcloud.generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file('img/us_by_category.png')
```
